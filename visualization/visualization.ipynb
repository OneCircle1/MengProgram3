```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
```

```python
df = pd.read_csv('ge.us.txt', sep=",", header=0)
df.columns = ["Date", "Open", "High", "Low", "Close", "Volume", "OpenInt"]
```

```python
df.head()
```

```python
plt.figure(figsize=(16,8))
plt.title('Highest Price History')
plt.plot([dt.datetime.strptime(d,'%Y-%m-%d').date() for d in df['Date']], df['High'])
plt.xlabel('Date')
plt.ylabel('Highest Price')
plt.savefig('High.png')
# plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('Lowest Price History')
plt.plot([dt.datetime.strptime(d,'%Y-%m-%d').date() for d in df['Date']], df['Low'])
plt.xlabel('Date')
plt.ylabel('Lowest Price')
plt.savefig('Low.png')
# plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('Volume History')
plt.plot([dt.datetime.strptime(d,'%Y-%m-%d').date() for d in df['Date']], df['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.savefig('Volume.png')
plt.show()
```
